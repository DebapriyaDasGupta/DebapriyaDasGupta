```python
import matplotlib.pyplot as plt

# Credits: https://pydicom.github.io/pydicom/stable/auto_examples/input_output/plot_read_dicom.html#sphx-glr-auto-examples-input-output-plot-read-dicom-py
if 'PixelData' in ds:
    rows = int(ds.Rows)
    cols = int(ds.Columns)
    print("Image size.......: {rows:d} x {cols:d}, {size:d} bytes".format(
        rows=rows, cols=cols, size=len(ds.PixelData)))
    if 'PixelSpacing' in ds:
        print("Pixel spacing....:", ds.PixelSpacing)
        
# Pixel data can be used to plot the CT scan       
```

```python
# plot the image using matplotlib
plt.imshow(ds.pixel_array, cmap=plt.cm.bone) 
plt.show()
# We see the cervical CT scan of the patient.
```

```python
import os
def read_all_dcm(PatientID):
    Path = "../input/rsna-2022-cervical-spine-fracture-detection/train_images/" + PatientID + "/"
    path, dirs, files = next(os.walk(Path))
    file_count = 10
    PathList = [0]*file_count
    for i in range(len(PathList)):   
        try:
            PathList[i+1] = Path + str(i+1) + ".dcm"
            plt.imshow(pydicom.dcmread(PathList[i+1]).pixel_array, cmap=plt.cm.bone)
            plt.show()
            print(PathList[i+1])
        except:
            pass
        
    return
```

```python
read_all_dcm("1.2.826.0.1.3680043.10001") # Below you can see all CT scans of the patient selected.
```
