## Group by

```python
group_train = train_modified[['date_block_num', 'shop_id', 'item_id', 'item_price', 'item_cnt_day']].groupby(['date_block_num', 'shop_id', 'item_id', 'item_price']).sum()
group_train.rename(columns = {'item_cnt_day':'item_cnt_month'}, inplace = True)
group_train
```

```python
inner_join = pd.merge(train_modified, test, on = ["shop_id", "item_id"], how = "inner")
inner_join
```

```python
inner_join.drop_duplicates(subset = ['shop_id', 'item_id', 'item_price'], inplace = True)
```

```python
inner_join[inner_join.duplicated()]
```
