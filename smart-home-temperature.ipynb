```python
train = pd.read_csv('../input/solar/smart-homes-temperature-time-series-forecasting/train.csv')
train.head()
```

```python
X_train = train.columns.drop(['Indoor_temperature_room', 'Date'])
y_train = train['Indoor_temperature_room']
print(type(X_train))
```

```python
# Encoding all the predictor variables to convert the categorical values to numerical values.
X_train_encoded = pd.get_dummies(train[X_train])
print("Total number of predictors after encoding = ", len(X_train_encoded.columns))

# Printing the list of columns after encoding to understand the encoding process
print(X_train_encoded.columns)
X_train_encoded
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train_encoded, y_train)
```

```python
#Output:
print("Intercept = ", reg.intercept_)
print("Coefficient = ", reg.coef_)
print(f"Equation:  Y = {reg.coef_} X + {reg.intercept_}")
```

## Testing

```python
test = pd.read_csv('../input/solar/smart-homes-temperature-time-series-forecasting/test.csv')
test.head()
```

```python
X_test = test.columns.drop(['Date'])
print(X_test)
print(type(X_test))
```

```python
# Encoding all the predictor variables to convert the categorical values to numerical values.
X_test_encoded = pd.get_dummies(test[X_test])
print("Total number of predictors after encoding = ", len(X_test_encoded.columns))

# Printing the list of columns after encoding to understand the encoding process
print(X_test_encoded.columns)
X_test_encoded
```

```python
y_pred = reg.predict(X_test_encoded)
output_df = pd.DataFrame(
    {      'Predicted': y_pred    }
)
print(output_df)
```
